```python
import matplotlib.pyplot as plt

def draw_3d_scenario(ax, p, l1, l2, title):

    # ---------- 점, 라벨 ----------
    ax.scatter(*p,  color='red', s=70)
    ax.text(p[0], p[1], p[2],
            f'p ({p[0]},{p[1]},{p[2]})',
            fontsize=10, ha='right', va='bottom')

    for pt, name in zip([l1, l2], ['l1', 'l2']):
        ax.scatter(*pt, color='red', s=45)
        ax.text(pt[0], pt[1], pt[2],
                f'{name} ({pt[0]},{pt[1]},{pt[2]})',
                fontsize=9, ha='right', va='bottom')

    # ---------- 선분 ----------
    ax.plot([l1[0], l2[0]],
            [l1[1], l2[1]],
            [l1[2], l2[2]],
            color='steelblue', linewidth=2)

    # ---------- 축, 라벨 ----------
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)

    # ---------- 축 범위 ----------
    ax.set_xlim(-0.2, 3.5)
    ax.set_ylim(-0.2, 2.5)
    ax.set_zlim(-0.2, 2.5)



# ----------------- 전체 Figure -----------------
fig = plt.figure(figsize=(14, 6))

# ---------- 상황 1 ----------
ax1 = fig.add_subplot(121, projection='3d')
p1 = [3, 1, 2]          # 점 p
l1 = [0, 0, 0]          # 선분 시작점
l2 = [2, 2, 1]          # 선분 끝점
draw_3d_scenario(ax1, p1, l1, l2, "Scenario 1 : p (3,1,2)")

# ---------- 상황 2 ----------
ax2 = fig.add_subplot(122, projection='3d')
p2 = [1, 1, 1]          # 점 p
# l1, l2는 동일하게 유지
draw_3d_scenario(ax2, p2, l1, l2, "Scenario 2 : p (1,1,1)")

# 시점 재설정 (tight_layout이 시점을 초기화할 수 있어서)
ax1.view_init(elev=40, azim=-70)
ax2.view_init(elev=40, azim=-70)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def draw_triangle_and_point(ax, p, t1, t2, t3, title):
    
    # ---------- 점 p 표시 ----------
    ax.scatter(*p, color='red', s=100)
    ax.text(p[0], p[1], p[2],
            f'p ({p[0]},{p[1]},{p[2]})',
            fontsize=10, ha='right', va='bottom')

    # ---------- 삼각형 꼭짓점 표시 ----------
    triangle_points = [t1, t2, t3]
    for pt, name in zip(triangle_points, ['t1', 't2', 't3']):
        ax.scatter(*pt, color='blue', s=70)
        ax.text(pt[0], pt[1], pt[2],
                f'{name} ({pt[0]},{pt[1]},{pt[2]})',
                fontsize=9, ha='right', va='bottom')

    # ---------- 삼각형 면 표시 ----------
    triangle = [[t1, t2, t3]]
    poly = Poly3DCollection(triangle, alpha=0.3, linewidths=1, edgecolors='blue', facecolor='cyan')
    ax.add_collection3d(poly)

    # ---------- 축, 라벨 ----------
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)

    # ---------- 축 범위 ----------
    ax.set_xlim(-0.2, 2.2)
    ax.set_ylim(-0.2, 2.2)
    ax.set_zlim(-0.2, 2.2)


# ----------------- 전체 Figure -----------------
fig = plt.figure(figsize=(8, 6))

# ---------- 상황 1 ----------
ax1 = fig.add_subplot(111, projection='3d')
p1 = [1, 0, 1]          # 점 p
t1_1 = [0, 0, 0]        # 삼각형 꼭짓점 1
t1_2 = [0, 1, 2]        # 삼각형 꼭짓점 2
t1_3 = [0, 2, 0]        # 삼각형 꼭짓점 3
draw_triangle_and_point(ax1, p1, t1_1, t1_2, t1_3, "Triangle & Point")

# 시점 재설정
ax1.view_init(elev=40, azim=-70)

plt.tight_layout()
plt.show()
```
