```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# ── Seaborn 테마 적용 ──────────────────────────────────────
sns.set_theme(style="white", font_scale=1.1)

fig = plt.figure(figsize=(18, 6))

# ── 색상 팔레트 (seaborn tab10 기반) ──────────────────────
palette = sns.color_palette("tab10", 5)
c1, c2, c3, c4, c5 = palette[0], palette[2], palette[1], palette[4], palette[3]
user_colors = [c1, c2, c3, c4, c5]


# ═══════════════════════════════════════════════════════════
# (a) FDMA — 주파수 분할 : 사용자별 수평 막대
# ═══════════════════════════════════════════════════════════
ax1 = fig.add_subplot(1, 3, 1)
ax1.set_facecolor('white')

users      = ['User 1', 'User 2', 'User 3', 'User 4', 'User 5']
y_positions = [4, 3, 2, 1, 0]
bar_height  = 0.75

for i, (user, y) in enumerate(zip(users, y_positions)):
    rect = mpatches.FancyBboxPatch(
        (0, y - bar_height / 2), 5, bar_height,
        boxstyle="round,pad=0.02",
        facecolor=user_colors[i], edgecolor='white', linewidth=1.5
    )
    ax1.add_patch(rect)
    ax1.text(2.5, y, user, ha='center', va='center',
             color='white', fontsize=12, fontweight='bold')

ax1.set_xlim(-0.3, 5.8);  ax1.set_ylim(-0.7, 4.7)
ax1.set_xticks([]);        ax1.set_yticks([])
ax1.spines[['top', 'right']].set_visible(False)

# 화살표 축
# ax1.annotate('', xy=(5.7, -0.65),   xytext=(0, -0.65),
#              arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
# ax1.annotate('', xy=(-0.25, 4.6),   xytext=(-0.25, -0.65),
#              arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax1.text(5.75, -0.65, 'Time',      ha='left',   va='center', fontsize=11)
ax1.text(-0.25, 4.7,  'Frequency', ha='center', va='bottom', fontsize=11)
ax1.set_title('(a) Frequency Division', fontsize=12, pad=10, fontweight='bold')


# ═══════════════════════════════════════════════════════════
# (b) TDMA — 시간 슬롯 분할 : 사용자별 수직 막대 + 반복
# ═══════════════════════════════════════════════════════════
ax2 = fig.add_subplot(1, 3, 2)
ax2.set_facecolor('white')

slot_users = [('User 1', c1), ('User 2', c2),
              ('User 3', c3), ('User 4', c4)]
slot_width = 0.85
slot_gap   = 0.15
bar_h      = 4.0
slot_labels = ['Slot 1', 'Slot 2', 'Slot 3', 'Slot 4']

for i, (user, color) in enumerate(slot_users):
    x_start = i * (slot_width + slot_gap)
    rect = mpatches.FancyBboxPatch(
        (x_start, 0), slot_width, bar_h,
        boxstyle="round,pad=0.02",
        facecolor=color, edgecolor='white', linewidth=1.5
    )
    ax2.add_patch(rect)
    ax2.text(x_start + slot_width / 2, bar_h / 2, user,
             ha='center', va='center', color='white', fontsize=11, fontweight='bold')

    # 슬롯 구간 표시 (상단 양방향 화살표)
    mid_x = x_start + slot_width / 2
    ax2.annotate('', xy=(x_start + slot_width - 0.05, bar_h + 0.3),
                 xytext=(x_start + 0.05, bar_h + 0.3),
                 arrowprops=dict(arrowstyle='<->', color='gray', lw=1.2))
    ax2.text(mid_x, bar_h + 0.55, slot_labels[i],
             ha='center', va='bottom', fontsize=9.5, color='gray')

# 점선 + User 1 반복 블록
x_repeat_start = 4 * (slot_width + slot_gap) + 0.4
ax2.plot([4 * (slot_width + slot_gap) + 0.05, x_repeat_start],
         [bar_h / 2, bar_h / 2], 'k--', lw=1.2)

rect_rep = mpatches.FancyBboxPatch(
    (x_repeat_start, 0), slot_width, bar_h,
    boxstyle="round,pad=0.02",
    facecolor=c1, edgecolor='white', linewidth=1.5
)
ax2.add_patch(rect_rep)
ax2.text(x_repeat_start + slot_width / 2, bar_h / 2, 'User 1',
         ha='center', va='center', color='white', fontsize=11, fontweight='bold')
ax2.text(x_repeat_start + slot_width / 2, bar_h + 0.55, '(repeat)',
         ha='center', va='bottom', fontsize=9, color='gray')

total_w = x_repeat_start + slot_width
ax2.set_xlim(-0.5, total_w + 0.5);  ax2.set_ylim(-0.8, bar_h + 1.0)
ax2.set_xticks([]);                  ax2.set_yticks([])
ax2.spines[['top', 'right']].set_visible(False)

# ax2.annotate('', xy=(total_w + 0.45, -0.75), xytext=(-0.1, -0.75),
#              arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
# ax2.annotate('', xy=(-0.45, bar_h + 0.95), xytext=(-0.45, -0.75),
#              arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax2.text(total_w + 0.5, -0.75, 'Time',      ha='left',   va='center', fontsize=11)
ax2.text(-0.45, bar_h + 1.0,   'Frequency', ha='center', va='bottom', fontsize=11)
ax2.set_title('(b) Time Division (TDMA)', fontsize=12, pad=10, fontweight='bold')


# ═══════════════════════════════════════════════════════════
# (c) CDMA — 3D 코드 분할 큐브
# ═══════════════════════════════════════════════════════════
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.set_facecolor('white')

def draw_cube_edges(ax, xlim, ylim, zlim, color='gray', lw=0.8, alpha=0.4):
    x0, x1 = xlim; y0, y1 = ylim; z0, z1 = zlim
    edges = [
        [(x0,y0,z0),(x1,y0,z0)], [(x0,y1,z0),(x1,y1,z0)],
        [(x0,y0,z1),(x1,y0,z1)], [(x0,y1,z1),(x1,y1,z1)],
        [(x0,y0,z0),(x0,y1,z0)], [(x1,y0,z0),(x1,y1,z0)],
        [(x0,y0,z1),(x0,y1,z1)], [(x1,y0,z1),(x1,y1,z1)],
        [(x0,y0,z0),(x0,y0,z1)], [(x1,y0,z0),(x1,y0,z1)],
        [(x0,y1,z0),(x0,y1,z1)], [(x1,y1,z0),(x1,y1,z1)],
    ]
    for e in edges:
        ax.plot([e[0][0],e[1][0]], [e[0][1],e[1][1]], [e[0][2],e[1][2]],
                color=color, lw=lw, alpha=alpha)

draw_cube_edges(ax3, [0,5], [0,5], [0,3])

# 사용자별 레이어 (아래→위: C, B, A)
layer_colors = [
    (*sns.color_palette("tab10")[2], 0.55),  # User C — green
    (*sns.color_palette("tab10")[1], 0.55),  # User B — orange
    (*sns.color_palette("tab10")[0], 0.55),  # User A — blue
]
layer_z      = [0.0, 1.0, 2.0]
layer_dz     = 0.85
user_names_3d = ['User C', 'User B', 'User A']

for z0, color, name in zip(layer_z, layer_colors, user_names_3d):
    verts = [
        [(0,0,z0),(5,0,z0),(5,5,z0),(0,5,z0)],
        [(0,0,z0+layer_dz),(5,0,z0+layer_dz),(5,5,z0+layer_dz),(0,5,z0+layer_dz)],
        [(0,0,z0),(5,0,z0),(5,0,z0+layer_dz),(0,0,z0+layer_dz)],
        [(0,5,z0),(5,5,z0),(5,5,z0+layer_dz),(0,5,z0+layer_dz)],
        [(0,0,z0),(0,5,z0),(0,5,z0+layer_dz),(0,0,z0+layer_dz)],
        [(5,0,z0),(5,5,z0),(5,5,z0+layer_dz),(5,0,z0+layer_dz)],
    ]
    poly = Poly3DCollection(verts, alpha=color[3],
                            facecolor=color[:3], edgecolor='white', linewidth=0.5)
    ax3.add_collection3d(poly)
    ax3.text(0.2, 0.3, z0 + layer_dz / 2, name,
             color='white', fontsize=10.5, fontweight='bold',
             ha='left', va='center', zorder=10)

ax3.set_xlim(0,5); ax3.set_ylim(0,5); ax3.set_zlim(0,3)
ax3.set_xticks([]); ax3.set_yticks([]); ax3.set_zticks([])
ax3.set_xlabel('Time',      fontsize=11, labelpad=4)
ax3.set_ylabel('Frequency', fontsize=11, labelpad=4)
ax3.set_zlabel('Code',      fontsize=11, labelpad=4)
ax3.xaxis.pane.fill = False
ax3.yaxis.pane.fill = False
ax3.zaxis.pane.fill = False
ax3.xaxis.pane.set_edgecolor('lightgray')
ax3.yaxis.pane.set_edgecolor('lightgray')
ax3.zaxis.pane.set_edgecolor('lightgray')
ax3.grid(False)
ax3.view_init(elev=22, azim=-50)
ax3.set_title('(c) Code Division (CDMA)', fontsize=12, pad=8, fontweight='bold')


# ── 전체 타이틀 및 저장 ────────────────────────────────────
# plt.suptitle('Multiple Access Resource Allocation Schemes',
#              fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout(pad=2.5)
plt.savefig('resource_allocation.png', dpi=150,
            bbox_inches='tight', facecolor='white')
plt.close()
```
