```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

correlations = [0.99, 0.75, 0.5, 0.25, 0, -0.25, -0.5, -0.75, -0.99]
n_points = 125

colors = [
    '#d62728',  # r=+0.99  진한 빨강
    '#e07b39',  # r=+0.75  주황
    '#e8b84b',  # r=+0.50  황금
    '#8db56b',  # r=+0.25  연두
    '#2ca02c',  # r= 0.00  진한 초록
    '#5aadbb',  # r=-0.25  청록
    '#4a90d9',  # r=-0.50  하늘
    '#3457b8',  # r=-0.75  파랑
    '#1a1aad',  # r=-0.99  진한 남색
]

fig, axes = plt.subplots(3, 3, figsize=(9, 9))
fig.patch.set_facecolor('white')
plt.subplots_adjust(wspace=0.1, hspace=0.2)

np.random.seed(42)

for idx, (corr, color) in enumerate(zip(correlations, colors)):
    row, col = divmod(idx, 3)
    ax = axes[row, col]

    mean = [0, 0]
    cov = [[1, corr], [corr, 1]]
    x, y = np.random.multivariate_normal(mean, cov, n_points).T

    ax.scatter(x, y,
               color=color,
               alpha=0.85,
               s=30,
               edgecolors='white',
               linewidths=0.3)

    ax.set_xlabel(f'r = {corr:+.2f}', fontsize=10, color='black', labelpad=6)
    ax.set_facecolor('white')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(-3.5, 3.5)
    ax.set_ylim(-3.5, 3.5)
    for spine in ax.spines.values():
        spine.set_edgecolor('#cccccc')
        spine.set_linewidth(0.8)

plt.savefig('correlation_scatter_3x3.png',
            dpi=150,
            bbox_inches='tight',
            facecolor='white')
plt.show()
```
